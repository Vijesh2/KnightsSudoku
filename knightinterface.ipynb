```python
# STEP 1a: some preliminary importing of libraries
from sklearn.cluster import KMeans
import numpy as np

# STEP 1b: specifying formatting for pretty-printing matrix values
float_formatter = lambda x: "%.2f" % x
np.set_printoptions(formatter={'float_kind':float_formatter})
```

```python
# STEP 2: load the data into a variable called allCustomers
# fields represent: 
# average_balance, aveMonthlySalary, numberOfATMvisiterPerMonth, numberDirectDebits
person1 = [1000, 2000, 12, 4]
person2 = [2200, 2500, 25, 12]
person3 = [3500, 5000, 8, 8]
person4 = [4500, 10000, 25, 5]
person5 = [5500, 5000, 20, 10]
person6 = [1500, 1500, 8, 16]
person7 = [2200, 800, 4, 10]
person8 = [3300, 1600, 5, 13]
person9 = [4400, 2200, 7, 15]
person10 = [3500, 2500, 10, 10]
allCustomers = np.array([person1, person2, person3, person4, person5, person6, person7, person8, person9, person10])
```

```python
# STEP 3: run the clustering algorithm
# note: requires user to specify how many clusters required
# choice of clusters is subjective. See: https://www.quora.com/How-can-we-choose-a-good-K-for-K-means-clustering
numberOfClusters = 4
kmeans=KMeans(n_clusters=numberOfClusters, random_state=0).fit(allCustomers)
```

```python
# STEP 4: show which segment each customer has been allocated to
kmeans.labels_
```

```python
# STEP 5: show where the 'centre' of each segment expressed as a value of each of the data fields e.g. average_balance
centers = kmeans.cluster_centers_
print (centers)
```

```python
# STEP 6: explore average for each data field within each segment
print ("The averages for allCustomers are {}".format(np.mean(allCustomers, axis=0)))
print ("The averages for each segment are {}".format(centers))
```

```python
# STEP 7: visualisation
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame(allCustomers)
axes = pd.tools.plotting.scatter_matrix(df, alpha=0.4, figsize=(20,10))
plt.tight_layout()
```
